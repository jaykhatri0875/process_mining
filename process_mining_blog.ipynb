```python
df = pd.DataFrame(sample_data,columns = ["Order_id","Event_type","Timestamp","done_by","Region"])
```

### Working with event logs

```python
event_log = pm4py.format_dataframe(df,case_id="Order_id",activity_key="Event_type", timestamp_key= "Timestamp", timest_format="%Y-%m-%d %H:%M:%S")
```

```python
start_act = pm4py.get_start_activities(event_log)
start_act
```

```python
end_act = pm4py.get_end_activities(event_log)
end_act
```

### Visualizing process graph

```python
dfg, start_activities, end_activities = pm4py.discover_dfg(event_log)
pm4py.view_dfg(dfg, start_activities, end_activities)
```

### Process stats

```python
performance_dfg, start_activities, end_activities = pm4py.discover_performance_dfg(event_log, case_id_key='case:concept:name', activity_key='concept:name', timestamp_key='time:timestamp')
```

```python
pm,start_activities, end_activities = pm4py.discover_petri_net_inductive(event_log)
pm4py.view_petri_net(pm,start_activities,end_activities)
```

```python
from pm4py.algo.discovery.alpha import algorithm as alpha_miner
net, initial_marking, final_marking = alpha_miner.apply(event_log)
```

```python
from pm4py.visualization.petri_net import visualizer as pn_visualizer
gviz = pn_visualizer.apply(net, initial_marking, final_marking)
pn_visualizer.view(gviz) 
```

```python
from pm4py.algo.discovery.inductive import algorithm as inductive_miner

from pm4py.algo.conformance.tokenreplay.diagnostics import duration_diagnostics

net, initial_marking, final_marking = inductive_miner.apply(event_log)

from pm4py.algo.conformance.tokenreplay import algorithm as token_based_replay
parameters_tbr = {token_based_replay.Variants.TOKEN_REPLAY.value.Parameters.DISABLE_VARIANTS: True, token_based_replay.Variants.TOKEN_REPLAY.value.Parameters.ENABLE_PLTR_FITNESS: True}
replayed_traces, place_fitness, trans_fitness, unwanted_activities = token_based_replay.apply(event_log, net, initial_marking, final_marking, parameters=parameters_tbr)

act_diagnostics = duration_diagnostics.diagnose_from_notexisting_activities(event_log, unwanted_activities)
for act in act_diagnostics:
    print(act, act_diagnostics[act])
```
